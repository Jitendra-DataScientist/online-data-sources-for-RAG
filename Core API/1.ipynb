```python
def extract_relevant_chunks(text, keywords, window_size=100):
    import re
    matches = []
    for keyword in keywords:
        for match in re.finditer(keyword, text, re.IGNORECASE):
            start = max(match.start() - window_size, 0)
            end = min(match.end() + window_size, len(text))
            snippet = text[start:end]
            matches.append(snippet)
    return matches

keywords = ['machine learning', 'algorithm', 'model']  # Replace with relevant keywords
relevant_chunks = extract_relevant_chunks(full_text, keywords)

for i, chunk in enumerate(relevant_chunks, 1):
    print(f"Chunk {i}:\n{chunk}\n{'-' * 40}")
```
