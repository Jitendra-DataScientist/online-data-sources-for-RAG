```python
import arxiv

# Create a default API client
client = arxiv.Client()

# Define the search query
search = arxiv.Search(
    query="quantum",
    max_results=10,
    sort_by=arxiv.SortCriterion.SubmittedDate
)

# Execute the search and print article titles
for result in client.results(search):
    print(result.title)
```

```python
import arxiv

# Specify the arXiv ID
arxiv_id = "1605.08386v1"

# Create a search object for the specific ID
search = arxiv.Search(id_list=[arxiv_id])

# Fetch the result
paper = next(arxiv.Client().results(search))

# Download the PDF to the current directory
paper.download_pdf()
```

```python
import arxivpy

# Define the search query for multiple categories
search_query = ['cs.CV', 'cs.LG', 'cs.CL', 'cs.NE', 'stat.ML']

# Fetch articles
articles = arxivpy.query(
    search_query=search_query,
    start_index=0,
    max_index=200,
    results_per_iteration=100,
    wait_time=5.0,
    sort_by='lastUpdatedDate'
)

# Download PDFs to the 'arxiv_pdf' directory
arxivpy.download(articles, path='arxiv_pdf')
```
